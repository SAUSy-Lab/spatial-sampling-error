```python
import numpy as np
import pandas as pd
import geopandas as gpd
import seaborn as sns
import geoplot
import shapely
import pysal
import matplotlib.pyplot as plt
%matplotlib inline
```

Create some random XY points between 0 and 100 in each dimension, and plot for fun

```python
# constat data gen variables
total_points = 10000
knn_for_data_gen = 30

# generate random location of points
xy = 1.2 * np.random.rand(total_points,2)
xydf = pd.DataFrame(data=xy)
xydf.columns = ['x', 'y']

# standard normal errors for each point
errors_ind = np.random.normal(0, 1, total_points)
xydf["errors"] = errors_ind

# an identiy matrix needed for generating simulated values at each point
I = np.identity(total_points)

# plotting the points
sns.relplot(x="x", y="y", data=xydf);
```

Generate a spatial weights matrix for the points based on $k$ nearest neighbours

```python
# weights matrix for k nearest
kd = pysal.lib.cg.kdtree.KDTree(xy)
W = pysal.lib.weights.KNN(kd, knn_for_data_gen)
W.transform = 'r' # row normalizing

# extract the sparse weights matrix as a full np array for matrix multiplication
W = (W.sparse)
W = (W.toarray())
```

Pick a $\rho$ for generating spatiall correlated values at each point, then run spatial autoregressive process

```python
rho = 0.9
```

$Y = (I - \rho W)^{-1} \epsilon$

```python
%%time 
Y = np.matrix(I - rho * W).I.dot(errors_ind)
```

```python
# append these Y values to the point data frame
xydf["sim"] = np.transpose(Y)
```

```python
# plot if we want to see the points with colours
sns.relplot(x="x", y="y", hue="sim",data=xydf);
```

```python
# also convert these Y values to a binary (1,0) if we want to analyze zonal proportions (where mean = 0.5)
simmean = xydf["sim"].mean()
xydf['sim_b'] = 0
xydf['sim_b'][xydf['sim'] > simmean] = 1
```

```python
# add in binomial distribution for whether observation is sampled or not, 
# do this for several sampling rates [0.03,0.05,0.1,0.2,0.5,1.0]
xydf["sample_03"] = np.random.binomial(1, 0.03, size=total_points)
xydf["sample_05"] = np.random.binomial(1, 0.05, size=total_points)
xydf["sample_10"] = np.random.binomial(1, 0.10, size=total_points)
xydf["sample_20"] = np.random.binomial(1, 0.20, size=total_points)
xydf["sample_50"] = np.random.binomial(1, 0.50, size=total_points)
xydf["sample_100"] = np.random.binomial(1, 1, size=total_points)
```

```python
# convert the points to a geopandas layer, for spatial joining
geometry = [shapely.geometry.Point(xy) for xy in zip(xydf.x, xydf.y)]
gdf = gpd.GeoDataFrame(xydf,  geometry=geometry)
```

```python
# load in grid data (was generated in QGIS by hand)
grid_6 = gpd.read_file("grids/grid_6x6.geojson")
grid_8 = gpd.read_file("grids/grid_8x8.geojson")
grid_10 = gpd.read_file("grids/grid_10x10.geojson")
grid_12 = gpd.read_file("grids/grid_12x12.geojson")
grid_15 = gpd.read_file("grids/grid_15x15.geojson")
```

```python
# spatial join the grid IDs to the point data
xy_with_grid_6 = gpd.sjoin(gdf, grid_6, how="inner", op='intersects')
xy_with_grid_8 = gpd.sjoin(gdf, grid_8, how="inner", op='intersects')
xy_with_grid_10 = gpd.sjoin(gdf, grid_10, how="inner", op='intersects')
xy_with_grid_12 = gpd.sjoin(gdf, grid_12, how="inner", op='intersects')
xy_with_grid_15 = gpd.sjoin(gdf, grid_15, how="inner", op='intersects')
```

```python
# generate means and proportions for each
grid_6_desc = xy_with_grid_6.groupby(['id']).agg({'errors': "count",'sim': "mean",  'sim_b': "sum"})
grid_8_desc = xy_with_grid_8.groupby(['id']).agg({'errors': "count",'sim': "mean",  'sim_b': "sum"})
grid_10_desc = xy_with_grid_10.groupby(['id']).agg({'errors': "count",'sim': "mean",  'sim_b': "sum"})
grid_12_desc = xy_with_grid_12.groupby(['id']).agg({'errors': "count",'sim': "mean",  'sim_b': "sum"})
grid_15_desc = xy_with_grid_15.groupby(['id']).agg({'errors': "count",'sim': "mean",  'sim_b': "sum"})
# update some of the column names
grid_6_desc["mean"] = grid_6_desc["sim"]
grid_6_desc["prop"] = grid_6_desc["sim_b"] / grid_6_desc["errors"]
del grid_6_desc['sim'], grid_6_desc['sim_b'], grid_6_desc['errors']
grid_8_desc["mean"] = grid_8_desc["sim"]
grid_8_desc["prop"] = grid_8_desc["sim_b"] / grid_8_desc["errors"]
del grid_8_desc['sim'], grid_8_desc['sim_b'], grid_8_desc['errors']
grid_10_desc["mean"] = grid_10_desc["sim"]
grid_10_desc["prop"] = grid_10_desc["sim_b"] / grid_10_desc["errors"]
del grid_10_desc['sim'], grid_10_desc['sim_b'], grid_10_desc['errors']
grid_12_desc["mean"] = grid_12_desc["sim"]
grid_12_desc["prop"] = grid_12_desc["sim_b"] / grid_12_desc["errors"]
del grid_12_desc['sim'], grid_12_desc['sim_b'], grid_12_desc['errors']
grid_15_desc["mean"] = grid_15_desc["sim"]
grid_15_desc["prop"] = grid_15_desc["sim_b"] / grid_15_desc["errors"]
del grid_15_desc['sim'], grid_15_desc['sim_b'], grid_15_desc['errors']
```

```python
# join this data to the grids for visualizing and computing spatial statistics
grid_6_join = grid_6.merge(grid_6_desc, on='id')
grid_8_join = grid_8.merge(grid_8_desc, on='id')
grid_10_join = grid_10.merge(grid_10_desc, on='id')
grid_12_join = grid_12.merge(grid_12_desc, on='id')
grid_15_join = grid_15.merge(grid_15_desc, on='id')
```

```python
# plot to visually see how clustered it is!
f, axarr = plt.subplots(1, 5, figsize=(12, 3))
geoplot.choropleth(
    grid_6_join, hue='mean',
    edgecolor='white', linewidth=1,
    cmap='Blues', legend=False,
    scheme='quantiles', figsize=(4, 5),
    ax=axarr[0]
)
geoplot.choropleth(
    grid_8_join, hue='mean',
    edgecolor='white', linewidth=1,
    cmap='Blues', legend=False,
    scheme='quantiles', figsize=(4, 5),
    ax=axarr[1]
)
geoplot.choropleth(
    grid_10_join, hue='mean',
    edgecolor='white', linewidth=1,
    cmap='Blues', legend=False,
    scheme='quantiles', figsize=(4, 5),
    ax=axarr[2]
)
geoplot.choropleth(
    grid_12_join, hue='mean',
    edgecolor='white', linewidth=1,
    cmap='Blues', legend=False,
    scheme='quantiles', figsize=(4, 5),
    ax=axarr[3]
)
geoplot.choropleth(
    grid_15_join, hue='mean',
    edgecolor='white', linewidth=1,
    cmap='Blues', legend=False,
    scheme='quantiles', figsize=(4, 5),
    ax=axarr[4]
)
```

```python
Wg6 = pysal.lib.weights.Queen.from_dataframe(grid_6_join)
Wg6.transform = 'r' # row normalizing
Wg8 = pysal.lib.weights.Queen.from_dataframe(grid_8_join)
Wg8.transform = 'r' # row normalizing
Wg10 = pysal.lib.weights.Queen.from_dataframe(grid_10_join)
Wg10.transform = 'r' # row normalizing
Wg12 = pysal.lib.weights.Queen.from_dataframe(grid_12_join)
Wg12.transform = 'r' # row normalizing
Wg15 = pysal.lib.weights.Queen.from_dataframe(grid_15_join)
Wg15.transform = 'r' # row normalizing
```

```python
mi6 = pysal.explore.esda.Moran(np.array(grid_6_join["prop"]), Wg6, two_tailed=False)
mi8 = pysal.explore.esda.Moran(np.array(grid_8_join["prop"]), Wg8, two_tailed=False)
mi10 = pysal.explore.esda.Moran(np.array(grid_10_join["prop"]), Wg10, two_tailed=False)
mi12 = pysal.explore.esda.Moran(np.array(grid_12_join["prop"]), Wg12, two_tailed=False)
mi15 = pysal.explore.esda.Moran(np.array(grid_15_join["prop"]), Wg15, two_tailed=False)
#Gi = pysal.explore.esda.G(np.array(grid_12_join["prop"]), Wg)
```

```python
print("6x6, I =", round(mi6.I,3), "p =", round(mi6.p_norm,3))
print("8x8, I =", round(mi8.I,3), "p =", round(mi8.p_norm,3))
print("10x10, I =", round(mi10.I,3), "p =", round(mi10.p_norm,3))
print("12x12, I =", round(mi12.I,3), "p =", round(mi12.p_norm,3))
print("15x15, I =", round(mi15.I,3), "p =", round(mi15.p_norm,3))
```

```python
YVar='mean'
XVars=['id']

grid_joins = [grid_6_join,grid_8_join,grid_10_join,grid_12_join,grid_15_join]
grid_names = ["6x6","8x8","10x10","12x12","15x15"]

outputs = []
c = 0
while c < len(grid_names):

    grid_join = grid_joins[c]
    
    # compute spatial weights matrix
    Wg = pysal.lib.weights.Queen.from_dataframe(grid_join)
    Wg.transform = 'r' # row normalizing
    
    mi = pysal.explore.esda.Moran(np.array(grid_join["prop"]), Wg, two_tailed=False)
    
    # xydf[[YVar]+XVars].corr() 
    Ym=grid_join[YVar].values.reshape((len(grid_join),1))
    Xm=grid_join[XVars].values
    mlag = pysal.model.spreg.ml_lag.ML_Lag(Ym,Xm,w=Wg,name_y=YVar, name_x=XVars)
    
    output = [grid_names[c], round(mi.I, 3),round(mi.p_norm, 3), mlag.rho,  mlag.z_stat[2][1]]
    outputs.append(output)
    
    c += 1
```
